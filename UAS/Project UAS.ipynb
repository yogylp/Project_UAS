```python
import cv2

img = cv2.imread('images.jpg') #imread = untuk membaca image
img_resize = cv2.resize(img, (540, 300))  #a,b. a = lebar b = tinngi
edged = cv2.Canny(img_resize, 330, 300) # a,b. a = kejelasan gambar(hitam, semakin banyak semakin tidak jelas)
cv2.imshow("Original Image",img_resize) # Sintaks: cv2.imshow (window_name, image)
# Parameter:
# window_name: Sebuah string yang mewakili nama jendela tempat gambar akan ditampilkan.
# image: Ini adalah gambar yang akan ditampilkan.
# Nilai Pengembalian: Tidak mengembalikan apa pun.

cv2.imshow("Edged Image", edged)

cv2.waitKey(0) # Parameter:	delay - Keterlambatan dalam milidetik. 0 adalah nilai spesial yang berarti "selamanya".
# Fungsi waitKey menunggu acara utama tanpa batas (ketika \ texttt {delay} \ leq 0) atau selama delaymilidetik, 
# ketika positif. 
# Karena OS memiliki waktu minimum antara beralih thread, 
# fungsi tidak akan menunggu persis delayms, 
# itu akan menunggu setidaknya delayms, tergantung pada apa lagi yang berjalan di komputer Anda pada waktu itu. 
# Ini mengembalikan kode tombol yang ditekan
# atau -1 jika tidak ada tombol yang ditekan sebelum waktu yang ditentukan telah berlalu.
cv2.destroyAllWindows() #Fungsi ini destroyAllWindowsmenghancurkan semua jendela HighGUI yang dibuka.
```

```python
import cv2
import numpy as np

img = cv2.imread('images.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gaussian = cv2.GaussianBlur(gray,(3,3),0) # memburamkan atau menghaluskan

#canny
img_canny = cv2.Canny(img,100,200)

#sobel
img_sobelx = cv2.Sobel(img_gaussian,cv2.CV_8U,1,0,ksize=5)
img_sobely = cv2.Sobel(img_gaussian,cv2.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely


#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]]) # Membuat matriks 1,1,1 merupakan baris [] kolom
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx) # Membuat img dengan kernelx
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)


cv2.imshow("Original Image", img)
cv2.imshow("Canny", img_canny)
cv2.imshow("Sobel X", img_sobelx)
cv2.imshow("Sobel Y", img_sobely)
cv2.imshow("Sobel", img_sobel)
cv2.imshow("Prewitt X", img_prewittx)
cv2.imshow("Prewitt Y", img_prewitty)
cv2.imshow("Prewitt", img_prewittx + img_prewitty)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
